```python
import csv
import pandas as pd
from pandas import DataFrame
import datetime
```

```python
df=pd.read_csv("ucl_data.csv",parse_dates=True)
print(df.head())
df['Start time'] = pd.to_datetime(df['Start time'])
df['End time'] = pd.to_datetime(df['End time'])
```

```python
curr=datetime.datetime(2011,11,28,0,0,0)
seq1 = []
seq1.append(curr)
while curr< datetime.datetime(2011,12,12,23,0,0):
    curr = curr + datetime.timedelta(minutes = 15)
    seq1.append(curr)

print(len(seq1))
```

```python
df2=pd.DataFrame(seq1,columns=['date time'])
print(df2.head())
```

```python
df2["Free"]=0
df2['month'] = df2['date time'].dt.month
df2['day'] = df2['date time'].dt.day
df2['hour'] = df2['date time'].dt.hour
df2['minu'] = df2['date time'].dt.minute



print(df2.head())
```

```python
df3=df.loc[df['Activity'] == 'Spare_Time/TV']
df3.head()
```

```python

print(df3['Start time'].dt.hour.head())
df3['Start time'].dt.minute.head()

```

```python
for i in range(len(df3)):
    mon=df3.iloc[i,0].month
    date=df3.iloc[i,0].day
    hou=df3.iloc[i,0].hour
    minu=df3.iloc[i,0].minute
    minu=minu-minu%15
    
    start=datetime.datetime(2011,mon,date,hou,minu,0)
    j=df2.index[df2['date time'] ==start ]
    flag=1
    
    while(flag==1):
        
        
        mon1=df2.iloc[j,2]
        date1=df2.iloc[j,3]
        hou1=df2.iloc[j,4]
        minu1=df2.iloc[j,5]
        
        mon2=df3.iloc[i,1].month
        date2=df3.iloc[i,1].day
        hou2=df3.iloc[i,1].hour
        minu2=df3.iloc[i,1].minute
    
    
        if( datetime.datetime(2011,mon1,date1,hou1,minu1) >  datetime.datetime(2011,mon2,date2,hou2,minu2) ):
            flag=0
        else:
            df2.iloc[j,1]=1
            
        j=j+1


print(df2.head())   
    
```
